## Text analytics


```python
# Taking input from user
sen = input("Enter sentence: ")

# Splitting sentence into words
words = sen.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Highest Salary from Employee Data


```python
# Employee dictionary
employees = {
    "vinu": 75000,
    "vishua": 68000,
    "vishu": 72000
}

# Finding highest salary
highest_employee = max(employees, key=employees.get)

# Printing result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

## Find Maximum and Minimum Values

```python
# Given list
numbers = [10,30,50,60,78,57]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## Count Products Above a Price Threshold

```python
# Given list of prices
prices = [4780, 1200, 99, 1500, 300]

# Initialize counter
count = 0

# Loop through each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

##  Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "A", "P"]

# Count present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

##  Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

## Count Character Frequency

```python
text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

## Calculate Average Marks

```python
marks = [70, 80, 65, 75]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)
```
